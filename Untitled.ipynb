```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Classifiers
DTC = DecisionTreeClassifier(max_depth=5)
GNB = GaussianNB()
KNC = KNeighborsClassifier(n_neighbors=1)

classifiers = [DTC,GNB,KNC]
```

```python
(Y,c) = make_moons(n_samples=300,noise=0.3, random_state=11)
plt.scatter(Y[:,0],Y[:,1],c=c)
```

```python
# CV
n = 300
n_fold= 5
fold = np.zeros((n,))
for i in range(n_fold):
    fold[(n*i)//n_fold:(n*(i+1))//n_fold] = i
```

```python
scores = np.zeros((n_fold,len(classifiers)))
for i in range(n_fold):
    for n_clf,clf in enumerate(classifiers):
        clf.fit(Y[fold==i],c[fold==i])
        score_fold = clf.score(Y,c)
        scores[i,n_clf]=score_fold
```

```python
scores
```

```python
def decision_boundaries(clf,X):
    h = .02
    x1_min, x1_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    x2_min, x2_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    x11, x22 = np.meshgrid(np.arange(x1_min, x1_max, h),np.arange(x2_min, x2_max, h))
    if hasattr(clf, "decision_function"):
        Z = clf.decision_function(np.c_[x11.ravel(), x22.ravel()])
    else:
        Z = clf.predict(np.c_[x11.ravel(), x22.ravel()])
    return Z
```

```python
plt.figure(figsize=(10,10))
for i in range(n_fold):
    subplot_idx = np.ravel_multi_index((i,0),(5,4))+1
    plt.subplot(n_fold,len(classifiers)+1,subplot_idx)
    plt.scatter(Y[fold==i,0],Y[fold==i,1],c=c[fold==i])
    if i==0:
        plt.title("Fold"+str(i))
    for j,clf in enumerate(classifiers):
        clf.fit(Y[fold==i],c[fold==i])
        Z=decision_boundaries(clf,Y)
        
        subplot_idx = np.ravel_multi_index((i,j+1),(5,4))+1
       # print(subplot_idx)
        
        plt.subplot(n_fold,len(classifiers)+1,subplot_idx)
        plt.contourf(x11, x22, Z.reshape(x11.shape), cmap=plt.cm.bwr, alpha=.8)
        if i==0:
            plt.title(str(clf).split("(")[0])
#plt.tight_layout()
```

```python
class fusion_vote_maj():
    def __init__(self,list_classifiers,is_train=False):
        self.list_classifiers=list_classifiers
        self.n_clf = len(list_classifiers)
        self.is_train = is_train

    def fit(self,X,y):
        if self.is_train:
            pass
        for i, clf in enumerate(self.list_classifiers):
            clf.fit(X,y)
        self.is_train=True

    def predict(self,X):
        assert self.is_train==True , "clf not trained"
        prediction = np.zeros(X.shape[0])
        for clf in self.list_classifiers:
            prediction += clf.predict(X)
        
        return (prediction/self.n_clf > 0.5).astype(int)
    
    def score(self,X,y):
        return (self.predict(X)==y).mean()
```
